```python
# importing all needed liabraries.

import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
```

# Data Collection and Pre-processing

```python
# loading the data
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a dataframe.
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the target column in the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of dataframe
data_frame.tail()
```

```python
# number of rows and columns in the datset
data_frame.shape
```

```python
# getting some information about data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of the target variable
data_frame['label'].value_counts()
```

# 1 --> Benign
# 0 --> Malignant

```python
data_frame.groupby('label').mean()
```

# Separating the features and target

```python
X = data_frame.drop(['label'],axis=1)
Y = data_frame['label']
```

```python
# print X
pd.DataFrame(X)
```

```python
# print Y because Y contains all the labels.
print(Y)
```

# Splitting the data into Training data and Testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print('The shape of the data is : {}'.format(X.shape))
print('The shape of the training data is : {}'.format(X_train.shape))
print('The shape of the testing data is : {}'.format(X_test.shape))
```

# Model training 


# Logistic Regression

```python
model = LogisticRegression(max_iter=10, random_state=42, verbose=0)
```

# Training the Logistic Regression model using training data


```python
model.fit(X_train, Y_train)
```

# Model Evaluation

# Accuracy Score

```python
# Accuracy on Training data model.
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data =',training_data_accuracy)
```

```python
# Accuracy on test data model.
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data model =',test_data_accuracy)
```
